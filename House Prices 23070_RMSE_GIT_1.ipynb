```python
df['SalePrice'].describe()
```

```python
object_columns = list(df.select_dtypes("object").columns)
```

```python
df_dum = pd.DataFrame()
```

```python
for column in object_columns:
    df_dum = pd.concat([df_dum, pd.get_dummies(df[column], prefix=column)], axis=1)
```

```python
# Создаем новый датафрейм, объединяя исходный датафрейм с новым. 
df_concat = pd.concat([df.select_dtypes(exclude='object'), df_dum], axis=1)
df_concat
```

```python
df_noid_price = df_concat.drop(['Id', "SalePrice"], axis=1)
df_noid_price
```

```python
target = 'SalePrice'
x = df_noid_price[:len(df_train['SalePrice'])].to_numpy()
y = df_concat[target][:len(df_train['SalePrice'])].to_numpy()
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split

x_train, x_valid, y_train, y_valid = train_test_split(x, y, test_size=0.35, random_state=1)
x_train.shape, x_valid.shape, y_train.shape, y_valid.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_norm = scaler.fit_transform(x_train)
x_valid_norm = scaler.transform(x_valid)
x_train_norm.shape, x_valid_norm.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


reg_1 = LinearRegression()
reg_1.fit(x_train, y_train)

# качество модели
print(f'''
    STD:
    train {y_train.std() :.2f}
    test {y_valid.std() :.2f}


    RMSE:
    train {mean_squared_error(reg_1.predict(x_train), y_train)**.5 :.2f}
    test {mean_squared_error(reg_1.predict(x_valid), y_valid)**.5 :.2f}

    R2:
    train {r2_score(reg_1.predict(x_train), y_train) :.2f}
    test {r2_score(reg_1.predict(x_valid), y_valid) :.2f}

''')


```

```python
from catboost import CatBoostRegressor
```

```python
cbr_1.fit(x_train_norm, y_train, verbose = 2)
y_pred_cbr = cbr_1.predict(x_valid_norm)
```
